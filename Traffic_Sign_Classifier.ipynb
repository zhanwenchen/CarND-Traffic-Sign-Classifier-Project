### Model Architecture

# LeNet-5
## Layer 1: Conv -> ReLU -> Pool
## Layer 2: Conv -> ReLU -> Pool
## Layer 3: FC -> ReLU
## Layer 4: FC -> ReLU
## Layer 5: FC

````python
# Use Tensorflow
import tensorflow as tf
from tensorflow.contrib.layers import flatten

# Parameters:
filter_size_width, filter_size_height = 3, 3 # standard
color_channels = 3
strides = [1, 1, 1, 1] # standard
padding = 'SAME' # Stanford cs231n recommends same padding over valid

layer1_filter_depth = 8
layer2_filter_depth = layer1_filter_depth * 2
layer3_output_depth = 400
layer4_output_depth = 120
layer5_output_depth = 43

# input_layer is a tf.placeholder, defined later before LeNet(input_layer) is invoked
def LeNet(input_layer):
    
    print("The shape of the input_layer is ", input_layer.get_shape())
    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # Layer 1: Convolutional. The output shape should be 32x32x32.
    # Activation using ReLU
    # Pooling using max pooling. The output shape should be 16x16x32.
    layer1_weights = tf.Variable(tf.truncated_normal(
        shape=(filter_size_width, filter_size_height, color_channels, layer1_filter_depth), 
        mean = mu, 
        stddev = sigma))
    layer1_biases = tf.Variable(tf.zeros(layer1_filter_depth))
    
    # Apply Convolution
    layer1 = tf.nn.conv2d(input_layer, layer1_weights, strides=strides, padding=padding)
    # Add bias
    layer1 = tf.nn.bias_add(layer1, layer1_biases)
    # Apply activation function ReLU
    layer1 = tf.nn.relu(layer1)
    # Apply max pooling
    layer1 = tf.nn.max_pool(
        layer1,
        ksize=[1, 2, 2, 1],
        strides=[1, 2, 2, 1],
        padding='VALID')
    
    print("The shape of layer 1's weights is ", layer1_weights.get_shape())
    print("The shape of layer 1's biases is ",layer1_biases.get_shape())
    print("The final shape of layer 1 is ", layer1.get_shape())
                                 
                                 
    # Layer 2: 
    # Convolutional. The output shape should be 16x16x64.
    # Activation using ReLU
    # Pooling. The output shape should be 8x8x64.
    layer2_weights = tf.Variable(tf.truncated_normal(
        shape=(filter_size_width, filter_size_height, layer1_filter_depth, layer2_filter_depth), 
        mean = mu, 
        stddev = sigma))
    layer2_biases = tf.Variable(tf.zeros(layer2_filter_depth))
    
    # Apply Convolution
    layer2 = tf.nn.conv2d(layer1, layer2_weights, strides=strides, padding=padding)
    # Add bias
    layer2 = tf.nn.bias_add(layer2, layer2_biases)
    # Apply activation function ReLU
    layer2 = tf.nn.relu(layer2)
    # Apply max pooling. The output shape should be 8x8x64
    layer2 = tf.nn.max_pool(
        layer2,
        ksize=[1, 2, 2, 1],
        strides=[1, 2, 2, 1],
        padding='VALID')
    
    layer2_shape = layer2.get_shape().as_list()
    print("The shape of the layer_2 output is ", layer2_shape)
    
    layer2_output_volume = layer2_shape[1] * layer2_shape[2] * layer2_shape[3]
    
    # Layer 3: Fully Connected.
    # Flatten. Flatten the output shape of the final pooling layer such that it's 1D instead of 3D. The easiest way to do is by using tf.contrib.layers.flatten, which is already imported for you.
    # ReLU
    layer3 = flatten(layer2) #output should be 4096
    
    print("layer 2's shape is ", tf.shape(layer3))
    layer3_weights = tf.Variable(tf.truncated_normal(
        shape=(layer2_output_volume, layer3_output_depth), 
        mean = mu, 
        stddev = sigma))
    layer3_biases = tf.Variable(tf.zeros(layer3_output_depth))
    # Add bias
    layer3 = tf.matmul(layer3, layer3_weights) + layer3_biases
    # ReLU activation
    layer3 = tf.nn.relu(layer3)
    
    print("The shape of the layer_3 output is ", layer3.get_shape())
    
    # Layer 4: FC -> ReLU
    layer4_weights = tf.Variable(tf.truncated_normal(
        shape=(layer3_output_depth, layer4_output_depth), 
        mean = mu, 
        stddev = sigma))
    layer4_biases = tf.Variable(tf.zeros(layer4_output_depth))
    # Add bias
    layer4 = tf.matmul(layer3, layer4_weights) + layer4_biases
    # ReLU activation
    layer4 = tf.nn.relu(layer4)
    

    print("The shape of the layer_4 output is ", layer4.get_shape())
    # Layer 5: Fully Connected (logits). Output size = 43
    layer5_weights = tf.Variable(tf.truncated_normal(
        shape=(layer4_output_depth, layer5_output_depth), 
        mean = mu, 
        stddev = sigma))
    layer5_biases = tf.Variable(tf.zeros(layer5_output_depth))
    # Add bias
    layer5 = tf.matmul(layer4, layer5_weights) + layer5_biases
    
    print("The shape of the layer_5 output is ", layer5.get_shape())
    
    print("logits is ", layer5)
    
    return layer5
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.

# 1. Features and Labels
# x is a placeholder for a batch of input images. y is a placeholder for a batch of output labels.
# You do not need to modify this section.

x = tf.placeholder(tf.float32, (None, image_shape[0], image_shape[1], color_channels))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, layer5_output_depth)

# 2. Training Pipeline
# You do not need to modify this section.

rate = 0.001

logits = LeNet(x)
print("LeNet finished running")
print("logits is ", logits)
# print("labels is ", labels)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)

# 3. Model Evaluation
# Evaluate how well the loss and accuracy of the model for a given dataset.
# You do not need to modify this section.
EPOCHS = 10
BATCH_SIZE = 128

correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples

# 4. Train the Model
# Before each epoch, shuffle the training set.
# After each epoch, measure the loss and accuracy of the validation set.
# Save the model after training.
# You do not need to modify this section.
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
    saver.save(sess, 'saved_models/lenet')
    print("Model saved")
    
    
# 5. Evaluate the Model
# Once you are completely satisfied with your model, evaluate the performance of the model on the test set.
# Be sure to only do this once!
# If you were to measure the performance of your trained model on the test set, then improve your model, and then measure the performance of your model on the test set again, that would invalidate your test results. You wouldn't get a true measure of how well your model would perform against real data.
# You do not need to modify this section.
with tf.Session() as sess:
    saver.restore(sess, tf.train.latest_checkpoint('saved_models/'))

    test_accuracy = evaluate(X_test, y_test)
    print("Test Accuracy = {:.3f}".format(test_accuracy))
````
